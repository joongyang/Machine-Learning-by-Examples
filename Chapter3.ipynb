## 그래프 작성

그래프를 작성하기 위해서는 `matplotlib` 라이브러리의 `pyplot` 모듈을 사용한다.
이 모듈에 대한 약칭은 `plt`로 하는 것이 관례이다.

작성한 그래프를 쥬피터 노트북에 나타나게 하려면 `%matplotlib inle`을 추가해야 한다.

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

그래프를 작성하기 위해서는 그래프 작성에 필요한 자료를 먼저 준비해야 한다.
대부분의 그래프는 2차원 평면 상에 점을 찍거나 이 점을 선으로 연결한 것이다.
따라서 2차원 평면 상에 점을 찍기 위해서 점의 가로 좌표와 세로 좌표가 준비되어야 한다.

$f(x) = (x-2) x (x+2)$의 그래프를 작성하고자 한다.

여러 $x$에서 $f(x)$를 계산해야 하므로 $f(x)$ 계산을 위한 함수를 정의하는 것이 좋다.

````python
# 리스트 2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return (x - 2) * x * (x + 2)
````

````python
# 리스트 2-(3)
print(f(np.array([1, 2, 3])))
````

$f(x) = (x-2)x(x+2)$에서 $(x-2)$의 2를 다른 값으로 하면 함수가 어떻게 변하는지도 알고 싶다면 함수의 정의를 다음과 같이 바꾸어야 한다.
즉, 2를 대체할 다른 값을 함수에 전달할 수 있게 새로운 매개변수를 도입해야 한다.

````python
# 리스트 2-(7)
# 매개변수가 두 개인 함수를 정의

def f2(x, w):
    return (x - w) * x * (x + 2)                             # (A) 함수 정의


# 함수 값을 계산할 x를 정의한다.
# 함수를 세밀하게 그리려면 함수 값을 조밀하게 계산해야 한다.

x = np.linspace(-3, 3, 100)                        # (B) [-3, 3] 구간을 100 등분

# 함수 그래프 작성하기

plt.plot(x, f2(x, 2), color='black', label='$w=2$')          # (C)
plt.plot(x, f2(x, 1), color='teal', label='$w=1$') # (D)
plt.legend(loc="upper left")                                 # (E) 범례 표시
plt.ylim(-15, 15)                                            # (F) 세로축의 범위
plt.xlim(-4, 4)                                              # 가로출 범위
plt.title('$f_2(x)$')                                        # (G) 제목
plt.xlabel('$x$')                                            # (H) 가로축 라벨
plt.ylabel('$y$', rotation=90) # rotation=레이블 회전 각도    # (I) 세로축 라벨
plt.grid(True)                                               # (J) 모눈 나타내기
plt.show()
````

## 3차원 그래프 그리기

$f(x_0, x_1) = (2x_0^2 + x_1^2) \exp (-(2x_0^2 + x_1^2) ) $의 그래프를 작성하기로 하자.

1. 그래프를 작성할 함수를 정의한다.
2. 함수를 계산한 여러 $x_0$, $x_1$ 를 배열로 생성한다.
3. 여러 $x_0$, $x_1$ 를 인자로 함수를 호출하여 함수 값을 계산한다.
4. 그래프를 작성한다.


````python
# 리스트 3-(1)
import numpy as np
import matplotlib.pyplot as plt

# 함수 f3을 정의
def f3(x0, x1):
    r = 2 * x0**2 + x1**2
    ans = r * np.exp(-r)
    return ans

# x0, x1에서 각각 f3을 계산
xn = 100                            # 함수를 계산할 x0, x1 값의 개수를 지정한다.
x0 = np.linspace(-2, 2, xn)       # (A) 함수를 계산할 x0의 값을 생성한다.
x1 = np.linspace(-2, 2, xn)       # (B) 함수를 계산할 x1의 값을 생성한다.
y = np.zeros((len(x0), len(x1)))  # (C) x0와 x1의 모든 조합에서 계산한 함수값을 저장할 배열을 만든다.
                                  # 가능한 모든 조합의 개수는 x0의 개수와 x1의 개수를 곱한 것이다.
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])  # (D) x0와 x1의 가능한 모든 조합에서 함수값을 계산하고 저장한다.
````

````python
# 리스트 3-(2)
print(x0)
````

````python
# 리스트 3-(3)
print(np.round(y, 1))
````

## 3차원 그래프를 입체적으로 그리기



````python
#리스트 3-(5)

# (A) 3차원 그래프를 그리기 위한 라이브러리
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1)      # (B) x0, x1의 모든 가능한 조합을 만들기 위해
                                    # x0의 행을 반복한 행렬, x1의 열을 반복한 행렬을 생성한다.
                                    # 생성된 두 행렬을 각 원소를 결합하면 한 가지 조합이 된다.

print(xx0)
print(xx1)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d')                     # (C) projection 매개변수에 '3d'를 지정한다.
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3,  # (D) 3축의 값을 이용하여 3차원 입체 그래프를 작성한다.
                color='blue', edgecolor='black') 
ax.set_zticks((0, 0.2))            # (E) 세 번째 축을 범위를 지정
ax.view_init(45, 45)              # (F) 3차원 그래프를 보는 시각을 지정한다
plt.show()
````

7.   scipy.stats.multivariate_normal 클래스를 이용하여 2변량 정규분포의 밀도함수 그래프를 등고선도로 작성하시오.



````python
import numpy as np
means = np.array([10, 20])  # 2변량 정규분포의 평균 벡터
cov_mat = np.array([[4, -4], [-4, 9] ])  # 2변량 정규분포의 공분산행렬
from scipy.stats import multivariate_normal
bi_normal = multivariate_normal(mean=means, cov=cov_mat) # 2변량 정규분포 객체 생성

# 2변량 정규분포의 밀도 계산
# 두 변수의 값을 3차원으로 준다.
# 다음은 두 변수의 값이 각각 (10, 19), (12, 19), (10, 21), (12, 21)인
# 4 점에서 밀도를 계산한다.
bi_normal.pdf([[[10, 12], [10, 12]], [[19, 19], [21, 21]]])  
````

````python
x1_lb = means[0] - 3 * np.sqrt(cov_mat[0][0])
x1_ub = means[0] + 3 * np.sqrt(cov_mat[0][0])
x2_lb = means[1] - 3 * np.sqrt(cov_mat[1][1])
x2_ub = means[1] + 3 * np.sqrt(cov_mat[1][1])

n = 100
x1 = np.linspace(x1_lb, x1_ub, n)
x2 = np.linspace(x2_lb, x2_ub, n)

xx1, xx2 = np.meshgrid(x1, x2)

# xx1 행렬, xx2 행렬 두개가 겹친 3차원 행렬을 만든다.
pos = np.empty(xx1.shape + (2,))  
pos[:, :, 0] = xx1               
pos[:, :, 1] = xx2
````

`scipy.stats.multivariate_normal` 객체의 `pdf()` 메소드로 밀도를 계산할 때는 변수의 값이 저장된 2차원 배열을 겹친 3차원 배열을 인자로 주어야 한다.

````python
pdf = bi_normal.pdf(pos)  # 밀도를 계산한다.
````

````python
plt.pcolor(x1, x2, pdf)
plt.colorbar()
````

````python
ax = plt.contour(x1, x2, pdf, levels=10)
ax.clabel(fmt='%4.2f')
plt.show()
````

````python
plt.pcolor(x1, x2, pdf, cmap='Blues')
plt.colorbar()
ax = plt.contour(x1, x2, pdf, levels=10)
ax.clabel(fmt='%4.2f')
plt.show()
````
