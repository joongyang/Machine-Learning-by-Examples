## 함수의 정의와 호출(실행)

````python
# 매개변수가 없는 함수의 정의
def my_func1():
    print('Hi!')
# 함수 my_func1()의 정의는 여기까지
my_func1() # 함수를 실행
````

````python
# 매개변수가 2개인 함수의 정의
# 매개변수는 함수를 호출하는 측이 함수에 데이터를 전달하는 통로이다
# return 문장은 함수가 함수를 호출하는 측에 데이터(실행 결과)를 전달한다.

def my_func2(a, b):
    c = a + b
    return c       # 반환값 또는 출력값(교재)라고 한다.


my_func2(1, 2)
````

return 문장으로 여러 개의 값을 반환하면 이 값을 요소로 하는 tuple이 생성되어 전달된다. (packing)

````python
def my_func3(D):
    m = np.mean(D)
    s = np.std(D)
    return m, s
````

문자열의 format() 메소드: 문자열 내에 지정한 값을 지정한 형식으로 출력해준다. 기본적으로 문자열 내의 {}에 해당하는 위치에 format() 메소드의 인자를 출력한다. {}에는 어떤 인자를 어떤 형식으로 출력할 것이지 기재한다.

````python
data=np.random.randn(100)
data_mean, data_std = my_func3(data)   # unpacking
print('mean:{0:3.2f}, std:{1:3.2f}'.format(data_mean, data_std))
# {3:5d}, {5:10s}
````

````python
output = my_func3(data)
print(output)
print(type(output))
print('mean:{0:3.2f}, std:{1:3.2f}'.format(output[0], output[1]))
````

## `numpy` 배열의 저장과 적재

`np.save()` 함수를 이용하여 `numpy` 배열을 파일에 저장하고, `np.load()` 함수를 이용하여 파일에 저장된 `numpy` 배열을 읽어 들일 수 있다.

`numpy` 배열을 저장한 파일의 기본 확장자는 `npy`이다.

````python
data = np.random.randn(5)
print(data)
np.save('datafile.npy', data)  # numpy 배열을 파일에 저장
data = []                      # 데이터 삭제(빈 list로 만듬), del data를 사용하면 객체를 삭제
print(data)
data = np.load('datafile.npy') # 파일에 저장된 numpy 배열을 프로그램 내로 읽어 들인다.
print(data)
````

`np.savez()` 함수를 이용하여 여러 개의 배열을 하나의 파일에 저장할 수 있으며, 한 파일에 저장된 여러 개의 배열을 분리하여 읽어 들일 수 있다.

여러 개의 배열을 하나의 파일에 저장할 때 확장자는 `npz`로 한다.

```python

np.savez('파일이름.npz', 파일내의배열이름=저장할배열이름, ...)
```

````python
data1 = np.array([1, 2, 3])
data2 = np.array([10, 20, 30])
np.savez('datafile2.npz', data1=data1, data2=data2) # 배열 data1, data2를 data1, data2라는 이름으로 파일에 저장한다.
data1 = [] # 데이터 삭제
data2 = []
outfile = np.load('datafile2.npz') # 파일에 저장된 모든 배열을 읽어 들인다.
print(outfile.files) # 읽어 들인 여러 배열의 이름을 조회할 수 있다.
data1 = outfile['data1'] # [] 연산자로 특정 배열을 지정할 수 있다.
data2 = outfile['data2'] # data2 꺼내기
print(data1)
print(data2)
````
